```python
def treat_outliers(data, column):
  plt.hist(data[column])
  plt.title(f'histogram {column}')
  plt.show()
  plt.violinplot(data[column])
  plt.title(f'violinplot{column}')
  plt.show()
  sns.boxplot(x=data[column])
  plt.title(f"Box Plot of {column} with Outliers")
  plt.show()
  Q1 = data[column].quantile(0.25)
  Q3 = data[column].quantile(0.75)
  IQR = Q3 - Q1
  lower_bound = Q1 - 1.5 * IQR
  upper_bound = Q3 + 1.5 * IQR
  outliers = data.loc[(data[column] < lower_bound) | (data[column] > upper_bound)]
  #data.loc[(data[column] < lower_bound) & (data[column]==lower_bound) | (data[column] > upper_bound) & (data[column]==upper_bound)]
  if (data[column]<lower_bound).any():
    data.loc[data[column]<lower_bound,column]=lower_bound
  # The following line was changed to use square brackets for indexing instead of parentheses
  if (data[column]>upper_bound).any():
    data.loc[data[column]>upper_bound,column]=upper_bound
  sns.boxplot(x=data[column])
  #plt.title(f"Box Plot of {column} After Outlier Treatment")
  #plt.show()
```
